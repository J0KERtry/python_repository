```python
import numpy as np 
import math
import copy
import random
import time
num = 16  # 数独のサイズを設定
del_count = 100 # 削除するマスの数を設定


"重複なしのランダムを生成し、リストで返す"
def rand_int(a, b, k):
    ns = []
    while len(ns) < k:
        n = random.randint(a, b)
        if not n in ns:
            ns.append(n)
    return ns

"対角線上のブロックにランダムな数字格納"
def prepare_block(num):
  list = [[0 for i in range(num)] for j in range(num)] # 数独用のパネル
  random = []   # 乱数を格納

  mass = int(math.sqrt(num))
  for i in range(mass):
    xbase = mass*i   # 対角線上の開始点が list[ybase][xbase]となる
    ybase = mass*i
    random = rand_int(1, num, num)
    for y in range(ybase, ybase + mass):
      for x in range(xbase, xbase + mass):
        list[y][x] = random[(y - ybase)*mass + (x - xbase)]

  return list


######################################
#数字を入れられるかチェックする関数群#
######################################
#横をチェック
def row(values, y, i, num):
    return all(True if i != values[y][_x] else False for _x in range(num))

#縦をチェック"
def column(values, x, i, num):    
    return all(True if i != values[_y][x] else False for _y in range(num))

#num x numのブロックをチェック"
def block(values, x, y, i, num):
    mass = int(math.sqrt(num))
    xbase = (x // mass) * mass
    ybase = (y // mass) * mass
    return all(True if i != values[_y][_x] else False
            for _y in range(ybase, ybase + mass)
                for _x in range(xbase, xbase + mass))

#すべてのチェックを満たしているかチェック
def check(values, x, y, i, num):
    return all([row(values, y, i, num), column(values, x, i, num), block(values, x, y, i, num)])


"入れられる数の少ないリスト番号を返す"
def count(values, num):
  min = num
  min_x, min_y = -1, -1   # 候補の個数が少ない座標を保存
  result = []   # 入る候補の数字を保存

  for y in range(num):
    for x in range(num):
      if values[y][x] == 0:  # マスが0のとき
        list = []
        for i in range(1, num+1):
          if check(values, x, y, i, num): # 入れられる数字を追加
            list.append(i)
        if (min > len(list)): # 入れられる候補の少ないものが見つかったら更新
          min = len(list)
          result = copy.copy(list)
          min_x, min_y = x, y

  return min_x, min_y, result

####################
## 数独を解く関数 ##
####################
def solver(values, x, y, num):
  min_x, min_y, list = count(values, num)  # 入れられる数の少ないインデックス取得

  if (min_x==-1): #終了条件
    return True
  for i in range(0, len(list)):
    values[min_y][min_x] = list[i]  #数字を入れる
    if solver(values, min_x, min_y, num):
      return True
    values[min_y][min_x] = 0 #戻ってきたら0に戻す    

  return False

########################################
## 完成している数独から要素を削除する ##
########################################
def draw_out(values, del_count):
  draw_out_index = []  
  draw_out_index = rand_int(0, num*num-1, del_count)  # 抜き出す配列の要素番号を取得
  for i in range(del_count):
    y = draw_out_index[i] // num
    x = draw_out_index[i] % num
    values[y][x] = 0


####################
####  実行部分  ####
####################
" 解けない問題であれば繰り返し生成 -> 解く "
start = time.time()   # 実行開始時間保存
data = prepare_block(num)   # 対角線上に数字が含まれているパネル(data)を作成
repeate = 1   # 問題を生成した回数
solver(data, 0, 0, num)   # 対角線上に数字が含まれているパネル(data)を作成
while True:
  if not all(True if 0 in data[_y] else False for _y in range(num)):  # 数独に0が含まれていない＝完成可能なら終了
    break
  else:  # 生成と解くを繰り返す
    data = prepare_block(num)
    repeate += 1
    solver(data, 0, 0, num)
performance_time = time.time() - start

print("数独を生成した回数：",repeate)
print("実行時間 : ", performance_time)
for a in data:
    print(*a)
print("\n")
draw_out(data, del_count)  #　要素を削除
print("生成した例題")
for a in data:
    print(*a)
```
